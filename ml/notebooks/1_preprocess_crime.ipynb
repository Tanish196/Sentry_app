```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\TRJ\\programming\\projects\\ml\\sentry\\Sentry_app\\ml\\datasets\\raw\\crime\\crime_delhi.csv")
district = pd.read_csv("../datasets\\raw\\crime\\delhi_ps_district_mapping.csv")
district['Police_Station_Area'] = district['Police_Station_Area'].str.strip().str.lower()
```

```python
df = df.merge(
    district, 
    left_on="nm_pol",
    right_on = "Police_Station_Area",
    how="left",
)
```

```python
df = df.drop(columns=['Police_Station_Area'])
```

```python
if "crime/area" not in df.columns:
    df["crime_per_area"] = df["totalcrime"] / (df["area"] + 1e-8)
else:
    df["crime_per_area"] = df["crime/area"]
```

```python
df.head()
```

```python
df["nm_pol"] = df["nm_pol"].str.strip().str.upper()

df = df[[
    "nm_pol",
    "totalcrime",
    "crime_per_area",
    "murder",
    "rape",
    "robbery",
    "theft",
    "District"
]]
```
