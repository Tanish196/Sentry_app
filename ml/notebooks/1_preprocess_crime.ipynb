```python
import pandas as pd
```

```python
df = pd.read_csv("..\\datasets\\crime\\crime_delhi.csv")
df2 = pd.read_csv("..\\datasets\\crime\\crime_delhi_filtered.csv")
```

```python
df2.columns
```

```python
df.columns
```

```python
df = df.drop(columns=['murder', 'rape', 'robbery', 'theft', 'assualt murders','sexual harassement', 'long', 'lat', 'crime/area'])
```

```python
df = df.merge(
    df2,
    left_on="nm_pol", 
    right_on="nm_pol",
    how="right"
)
```

```python
df['district'].unique()
```

```python
if "crime/area" not in df.columns:
    df["crime_per_area"] = df["totalcrime"] / (df["area"] + 1e-8)
else:
    df["crime_per_area"] = df["crime/area"]
```

```python
df.head()
```

```python
df["nm_pol"] = df["nm_pol"].str.strip().str.upper()
```

```python
len(df['nm_pol'].unique())
```

```python
df.head()
```

```python
def convert_district(cell_value, mapping_dic):
    return mapping_dic.get(cell_value, cell_value)

dic = {
    "Shahdara": "EAST",
    "Outer-North": "OUTER",
    "Rohini": "NORTH WEST"
}

df['district'] = df['district'].apply(lambda x: convert_district(x, dic))
df['district'] = df['district'].str.upper()
```

```python
df.to_csv("..\\datasets\\crime/crime_processed.csv", index=False)
```
