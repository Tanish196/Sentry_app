```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../datasets/accident/delhi_accident_data.csv")
```

```python
df.head()
```

```python
df['Traffic_Districts'].unique()
```

```python
df.columns
```

```python
years = [2018, 2019]
records = []

for _, row in df.iterrows():
    for year in years:
        records.append({
            "Traffic_Districts": row["Traffic_Districts"],
            "Year": year,
            "Fatal_count": row[f"Fatal_{year}"],
            "Injury_count": row[f"Injury_{year}"]
        })

traffic_df = pd.DataFrame(records)
```

```python
traffic_df
```

```python
OUTPUT_PATH = "..\\datasets\\accident\\accident_processed.csv"

traffic_df.to_csv(OUTPUT_PATH, index=False)
```
