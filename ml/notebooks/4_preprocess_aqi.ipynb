```python
import pandas as pd
import ast

INPUT_PATH = "..\\datasets\\raw\\aqi\\delhi_aqi_with_police_areas.csv"
OUTPUT_PATH = "..\\datasets\\processed\\aqi\\aqi_processed.csv"

df = pd.read_csv(INPUT_PATH)

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

```

```python
type(df["Police_Station_Areas"][0])
```

```python
def convert_to_list(x):
    y = list(map(str, x.split(',')))
    z = []
    for i in y:
        z.append(i.strip())
    return z
```

```python
df["Police_Station_Areas"] = df["Police_Station_Areas"].apply(convert_to_list)
```

```python
df["Police_Station_Areas"][0]
```

```python
df = df.explode("Police_Station_Areas")
```

```python
df
```

```python
df["Police_Station_Areas"] = (
    df["Police_Station_Areas"]
    .astype(str)
    .str.strip()
    .str.upper()
)

df = df[df["Police_Station_Areas"] != ""]

df = df.sort_values(["Police_Station_Areas", "Date"])

df["AQI"] = (
    df.groupby("Police_Station_Areas")["AQI"]
    .apply(lambda x: x.ffill().bfill())
    .reset_index(level=0, drop=True)
)

df["AQI"] = df["AQI"].fillna(df["AQI"].median())

df = df[[
    "Police_Station_Areas",
    "Date",
    "AQI"
]]

df.to_csv(OUTPUT_PATH, index=False)
```
