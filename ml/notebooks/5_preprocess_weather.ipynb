```python
import pandas as pd
```

```python

INPUT_PATH = "..\\datasets\\raw\\weather\\delhi_weather_2018_2019.csv"
OUTPUT_PATH = "..\\datasets\\processed\\weather\\weather_processed.csv"

df = pd.read_csv(INPUT_PATH)
```

```python
if {"Year", "Month", "Date"}.issubset(df.columns):
    df["Day"] = df["Date"]

    df["Date"] = pd.to_datetime(
        dict(
            year=df["Year"],
            month=df["Month"],
            day=df["Day"]
        ),
        errors="coerce"
    )

else:
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

df = df.dropna(subset=["Date"])

df = df.sort_values("Date")

```

```python
weather_cols = [
    "Avg Temperature",
    "Avg Humidity",
    "Total Precipitation"
]

for col in weather_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[weather_cols] = df[weather_cols].interpolate(method="linear")

for col in weather_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
print(df.isna().sum())
```

```python
df = df[["Date"] + weather_cols]
```

```python
df.to_csv(OUTPUT_PATH, index=False)
```
